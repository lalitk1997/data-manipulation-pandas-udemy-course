```python
#basic indexing, labelling and ordering
#loading libraries
import pandas as pd
df = pd.read_csv("./data/P87-S4-Basic-Data-Manipulations-Resources/AB_NYC_2019.csv")
df.head(3)
```

```python
#changes pandas default index column with id column
df2 = df.set_index("id")
df2.head(2)
```

```python
df3 = df.groupby("room_type").mean()
#now index is room-type
df3.head(2)
```

```python
#index is switched back to pandas default index
#but changes are not saved to df3
df3.reset_index()
#df3
```

```python
#type (index-column) is dropped
df3.reset_index(drop=True)
```

```python
#sorting is implemented on ondex only
df3.sort_index(ascending=False)
#df3.sort_index()
```
