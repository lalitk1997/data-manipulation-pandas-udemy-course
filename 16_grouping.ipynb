```python
#basic-grouping 
import pandas as pd
import numpy as np
df = pd.read_csv("./data/P87-S5-Grouping-Resources/train.csv", parse_dates=['Date'], low_memory=False)
```

```python
df.info()
```

```python
dfg = df.groupby("Store")
dfg.head()
```

```python
store_avg = dfg.mean()
store_avg.head(3)
```

```python
store_avg = store_avg.reset_index()
```

```python
store_avg.plot.scatter("Store", "Sales", s=1, title="Avg sale per store")
```
