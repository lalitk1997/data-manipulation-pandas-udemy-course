```python
#differences between numpy and pandas
# loading numpy and panda
import numpy as np
import pandas as pd
```

```python
#using pandas
df = pd.read_csv("data/P87-S2-Dataset-Basics-Resources/heart.csv")
df.head()
```

```python
#using numpy
#all values are converted to float
#2-d array representation
data = df.to_numpy()
print(data.dtype,data)
```

```python
#chnaging value in numpy
#value 63 changed to 0
data[0,0] = 0
print(data.dtype, data)
```

```python
#copying a particular segment from pandas dataframe to numpy
data = df[["age", "sex", "cp"]]
data2 = data.to_numpy()
print(data2.dtype, data2)
```

```python
#changing the values with 2-d array acess
data2[0,0] = 2
print(data2.dtype, data2)
```
