```python
#playing with NA/missing values, replacing etc 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("./data/P87-S4-Basic-Data-Manipulations-Resources/AB_NYC_2019.csv")
df.head(2)
```

```python
df.info()
#multiple NAN values exists
```

```python
#all rows having NA values in any column are dropped
df = df.dropna()
df.info()
```

```python
df.dropna(subset=["last_review"]).info()
df.info()
```

```python
#fill NA values with 0 integer
df.fillna(0).isna()
```

```python
#generic replace
#John is replaced with jono
df.replace("John", "Jono").head(1)
```

```python
df.host_name.replace("John", "Jono", limit=1)[0]
```

```python
#using matplotlib
plt.hist(df.price);
```

```python
plt.hist(df.price.clip(upper = 1000));
```

```python
df.loc[df.price > 1000, "price"] = 1000
plt.hist(df.price)
```
