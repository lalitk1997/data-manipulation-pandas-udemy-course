```python
#merging-types
#loading data
import pandas as pd

df_s = pd.read_csv("./data/P87-S6-Merging-Resources/students.csv")
df_p = pd.read_csv("./data/P87-S6-Merging-Resources/contact.csv")
df_g = pd.read_csv("./data/P87-S6-Merging-Resources/grades1.csv")
```

```python
df_g.tail()
```

```python
df_desn = df_g[df_g.course == "DESN101"]
df_desn.head(2)
print(df_desn.shape, df_s.shape)
#25 students but only 12 of them have taken desn101 course
```

```python
#merge-types
df_s.head(1) #id, firstname, lastname
df_desn.head(1) #student_id, course, grade
#combining both tables
pd.merge(df_s, df_desn, left_on="id", right_on="student_id").head(2)
```

```python
#left-join will contain all the student rows
df3 = pd.merge(df_s, df_desn, left_on="id", right_on="student_id", how="left");
df3
#print(df_s.size, df_desn.size)
#df_s
#df_desn
```

```python
#implementing outer join
df_a = pd.DataFrame({"A": ["x", "y", "z"], "B": [1, 2, 3]})
df_b = pd.DataFrame({"A": ["u", "v", "x"], "C": [5.0, 4.0, 3.0]})

display(df_a, df_b);
```

```python
#similar to cartesian cross product
pd.merge(df_a, df_b, left_on="A", right_on="A", how="outer")
```

```python
#inner join
pd.merge(df_a, df_b, on="A")
```

```python
pd.merge(df_a, df_b, on="A", how="outer")
```

```python
#duplicate-keys
#multiple values of x in column A in both dataframe
df_c = pd.DataFrame({"A": ["x", "x", "z"], "B": [1, 2, 3]})
df_d = pd.DataFrame({"A": ["u", "x", "x"], "C": [5.0, 4.0, 3.0]})

display(df_c, df_d);
```

```python
pd.merge(df_c, df_d, on="A")
```

```python
#Dublicate columns
df_e = pd.DataFrame({"A": ["x", "y", "z"], "B": [1, 2, 3]})
df_f = pd.DataFrame({"A": ["u", "v", "x"], "B": [5.0, 4.0, 3.0]})

display(df_e, df_f)
```

```python
pd.merge(df_e, df_f, on="A")
```

```python
df_m = pd.merge(df_e, df_f, on="A", suffixes=("_left", "_right"))
df_m
```

```python
df_m.rename(columns = {"A":"Alpha", "B_left":"Beta", "B_right":"Gamma"})
```

```python
#velidating expecting outputs
#one-to-one, one-to-many, many-to-one, many-to-many
pd.merge(df_s, df_p, left_on="id", right_on="student_id", validate="one_to_one")
```

```python
#row with index 0 is added one more time in the dataframe
df_p2 = df_p.append(df_p.iloc[0])
df_p2
```

```python
pd.merge(df_s, df_p2, left_on="id", right_on="student_id", validate="one_to_many")
```

```python
#composite keys
df_1 = pd.DataFrame({"year": [2000, 2000, 2001, 2001], "sem": [1, 2, 1, 2], 
                     "fee": [200, 200, 200, 200]})
df_2 = pd.DataFrame({"year": [2000, 2000, 2001, 2001], "sem": [1, 2, 1, 2], 
                     "student": [1, 2, 2, 3], "discount": [0.1, 0.2, 0.2, 1.0]})
df_3 = pd.DataFrame({"student": [1, 2, 3, 4, 5]})

display(df_1, df_2, df_3)
```

```python
#combining table based on multiple rows
combined = pd.merge(df_1, df_2, on=["year", "sem"])
combined["due"] = combined.fee * (1 - combined.discount)
combined
```

```python
pd.merge(df_3, df_2, on="student", how="left")
```
