```python
#basic plots
#importing lib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#loading csv
df = pd.read_csv("./data/P87-S3-Visual-Exploration-Resources/heart.csv")
```

```python
df.head()
```

```python
#reset_index will take the index as new column
chest_pain = df.groupby(by="cp").median().reset_index()
chest_pain.head()
```

```python
chest_pain.plot.bar(x="cp", y="age");
```

```python
chest_pain.plot.bar(x="cp")
```

```python
#matplotlib
fig, ax = plt.subplots()
ax.bar(chest_pain["cp"], chest_pain["age"], label="age")
ax.set_xlabel("cp")
ax.set_ylabel("age")
ax.legend();
```

```python
#set_xticks ticks at the perfect location of cp on x axis
fig, ax = plt.subplots()
ax.bar(chest_pain["cp"], chest_pain["age"], label="age",
       edgecolor = "k", color=["red", "green", "blue", "black"])
ax.set_xlabel("cp")
ax.set_ylabel("age")
ax.set_xticks(chest_pain["cp"]);
```

```python
#using seaborn
ax = sb.barplot(df["cp"], df["age"], errcolor="black", capsize=0.1)
```

```python
#scatter plots
df.plot.scatter("age", "trestbps")
```

```python
#s represent size and c represents the color within the marker
fig, ax = plt.subplots()
ax.scatter(df["age"], df["trestbps"], marker="o", s=100, c=df["age"],
          edgecolors="k", alpha=0.5)
ax.set_xlabel("age")
ax.set_ylabel("trestbps");
```

```python
#using seaborn
sb.scatterplot("age", "trestbps", hue="age", s=30, edgecolor="none", data=df)
```

```python
#line plots
ages = df.groupby("age").median().reset_index()
ages.head()
```

```python
ages.plot.line("age","chol")
```

```python
ages.plot.line("age",["chol", "trestbps"])
```

```python
#using matplotlib
#ls used for dotted line and lw is used for block size within the dotted line
fig, ax = plt.subplots()
ax.plot(ages["age"], ages["trestbps"], ls=":", lw=1.5)
ax.set_xlabel("age")
ax.set_label("trestbps")
```

```python
sb.lineplot("age", "trestbps", data=ages);
```
