```python
#adding and removing data
import pandas as pd
#loading dataset
df = pd.read_csv("./data/P87-S4-Basic-Data-Manipulations-Resources/astronauts.csv")
df.head(2)
```

```python
#conversion of data-types of columns 
#time-series, categoricals, strings to numeric
df["Birth Date"] = pd.to_datetime(df["Birth Date"])
df["Birth Date"]
```

```python
df["Birth Date"].dt.year.head(3)
```

```python
#subtracting to date-time data-types
zarya = pd.to_datetime("1998-11-20")
df["age_at_zarya"] = (zarya - df["Birth Date"]).astype("timedelta64[Y]")
df.head(3)
```

```python
df = df.drop(columns="birthdate")
df.info()
```

```python
#categoricals data 
df["Military Rank"].unique()
df.info() #military_rank column is object not categorical type
df["Military Rank"].dtype
```

```python
df["Military Rank"] = df["Military Rank"].astype("category")
df["Military Rank"].dtype
```

```python
#numeric/string conversion
df.head(1)
```

```python
df["age_at_zarya"].dtype
#column is float at first
#float can be converted to int directly
df["age_at_zarya"].astype("int");
#strings cannot be converted to int
df["age_at_zarya"].astype("str").astype(float).astype(int);
```
