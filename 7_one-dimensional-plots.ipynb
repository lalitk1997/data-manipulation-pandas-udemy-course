```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

df = pd.read_csv('./data/P87-S3-Visual-Exploration-Resources/heart.csv')
df.sample(4)
```

```python
#Histogram
df.age.plot.hist(bins = 40)
```

```python
plt.hist(df.age, bins=40);
```

```python
fig, ax = plt.subplots()
ax.hist(df.trestbps, bins=20, histtype="step", label="trestbps")
ax.hist(df.thalach, bins=20, histtype="stepfilled", label="thalach", alpha=0.3)
ax.legend()
```

```python
#boxplot
#mean, quartile and outliers IQR(inter-quartile range) = 75%-25%. wiskers-length =1.5 * IQR
df[["trestbps", "thalach"]].plot.box();
```

```python
df[["trestbps", "thalach"]].head()
df[["trestbps", "thalach"]].describe()
```

```python
#matplotlib
plt.boxplot(df[["trestbps", "thalach"]].to_numpy())
plt.xticks([1, 2], ("trestbps", "thalach"));
```

```python
#seaborn
sb.boxplot(data=df[["trestbps","thalach"]]);
```

```python
sb.boxplot(x="cp", y="trestbps", data=df)
```

```python
df.groupby("cp").boxplot(column="trestbps", sharex=True, layout=(1,4), grid=False)
plt.tight_layout();
```

```python
#vilon plots
fig , ax = plt.subplots()
ax.violinplot(df[["trestbps", "thalach"]].to_numpy());
```

```python
#bandwidth to control smoothing
fig, ax = plt.subplots()
ax.violinplot(df[["trestbps", "thalach"]].to_numpy(), bw_method=0.2);
```

```python
#violin plot with quartile range
sb.violinplot(data=df[["trestbps", "thalach"]], inner="quartile", bw=0.2)
```

```python
#be-swarm plots
sb.swarmplot(data=df[["trestbps", "thalach"]], size=3);
```

```python
#swarm + violin plot
sb.violinplot(data=df[["trestbps", "thalach"]], inner=None);
sb.swarmplot(data=df[["trestbps", "thalach"]], size=2, color="k", alpha=0.5)
```
