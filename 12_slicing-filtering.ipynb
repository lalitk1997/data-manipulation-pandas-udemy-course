```python
#basics-filtering-and-slicing
import pandas as pd
import numpy as np
#loading data
df = pd.read_csv("./data/P87-S4-Basic-Data-Manipulations-Resources/AB_NYC_2019.csv")
df.head(2)
```

```python
#slicing-columns
df["host_name"]
```

```python
df.host_name
```

```python
#extracting-two-column at a time
df[["host_name", "neighbourhood_group"]]
```

```python
#filtering on rows-(mask-filtering)
df[df.host_name =="Taz"].head(2)
```

```python
(df.host_name == "Taz").sum()
```

```python
mask = df.host_name == "Taz"
df[mask].head(2)
```

```python
quick_and_cheap = (df.price < 100) & (df.minimum_nights < 3)
quick_and_cheap.sum()
```

```python
df[quick_and_cheap].head(2)
```

```python
quick_and_cheap = (df.price < 100) & (df.minimum_nights < 3)
quick_and_cheap.sum()
```

```python
df[quick_and_cheap].head(2)
```

```python
reviews_mask = (df.reviews_per_month > 3) | (df.number_of_reviews > 50)
reviews_consistent = df[reviews_mask]
reviews_consistent.head(2)
```

```python
mask = np.logical_or((df.reviews_per_month > 3), (df.number_of_reviews > 50))
df[mask].head(2)
```

```python
df[~mask].head(2)
```

```python
#filtering rows and column together
#method used - .loc
df.loc[mask, ["name", "host_name"]]
```

```python
df.loc[mask, :].head(2)
```

```python
#filtering-based-on-index
df.iloc[0, :]
```

```python
df2 = df.set_index("id")
df2.iloc[0, :]
```

```python
#mask-helpers
df.loc[df.price.between(100, 200), "price"].head(2)
```

```python
df.loc[df.price.isin([100, 200]), "price"].head(2)
```

```python
mask_John = df["host_name"] == "John"
df.loc[mask_John, "host_name"].head()
```

```python
df == "John"
```

```python
#searching in columns
(df == "John").any()
```

```python
#searching in rows
(df == "John").any(axis=1)
```
